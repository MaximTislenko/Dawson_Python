## Quotation_Manipulation.py

```
# Манипуляции с цитатой
# Демонстрирует строковые методы

# цитата из Томаса Уотсона, который в 1943 году был директором IBM
quote = "Думаю, на мировом рынке можно будет продать штук пять компьютеров."

print("Исходная цитата:")
print(quote)

print("\nОна же в верхнем регистре:")
print(quote.upper())

print("\nВ нижнем регистре:")
print(quote.lower())

print("\nКак заголовок:")
print(quote.title())

print("\nС ма-а-аленькой заменой:")
print(quote.replace("штук пять", "несколько миллионов"))

print("\nОбращение")
print(quote.swapcase())

print("\ncapitalize():")
print(quote.capitalize())

print("\nstrip():")
print(quote.strip())

print("\nА вот опять исходная цитата:")
print(quote)

input("\n\nнажмите Enter, чтобы выйти.")

# upper()
# lower()
# swapcase() - регистр всех сиволов обращен
# capitalize() - первая буква в строке прописная, остальные строчные
# title()
# strip()
# replace(что заменить, на что заменить, [, сколько раз])
```

Программа должна включать в себя ряд дополнительных сборов, таких как налоги, лицензия, подготовка дилера и сбор за пункт назначения. Сделайте налог и лицензию процентом от базовой цены.
```
# Глава 2, Задача 4
# Напишите программу «Продавец автомобилей», в которой пользователь вводит базовую цену автомобиля.
# Программа должна включать в себя ряд дополнительных сборов, таких как налоги, лицензия, подготовка дилера
# и сбор за пункт назначения. Сделайте налог и лицензию процентом от базовой цены.
# Для остальных сборов должны быть установлены значения.
# Отображение фактической цены автомобиля после применения всех дополнительных услуг.

base_price = int(input("Какова базовая цена автомобиля? ₽"))

tax = int(0.20 * base_price)
lic = int(0.15 * base_price)
deal_prep = 5000
dest_charge = 4000

print("Налоги \t\t\t\t₽", tax, "\nПлата за лицензию \t\t₽", lic, "\nПодготовка дилера \t\t₽", \
      deal_prep, "\nПлата за пункт назначения \t₽", dest_charge, "\nИтого: \t\t\t\t₽", \
      int(base_price+tax+lic+deal_prep+dest_charge), sep='')

input("\n\nнажмите Enter, чтобы выйти.")
```
