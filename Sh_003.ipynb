```
# Текстовые задачи
# Демонстрирует математические операции

print("Если беременная самка бегемота массой 800 кг родит детеныша массой 40 кг,")
print("но затем съест 20 кг корма. То сколько она будет после всего этого весить?")
input("Чтобы узнать нажмите Enter.")
print("800 - 40 + 20 =", 800 - 40 + 20, "кг")

print("\nЕсли студент, возвращаясь с успешно сданного экзамена, на радостях")
print("купил каждому из 6 друзей по 3 бутылки пива. То сколько всего бутылок пива куплено?")
input("Чтобы узнать нажмите Enter.")
print("6 * 3 =", 6 * 3, "бутылок")

print("\nЕсли в ресторане вам с приятелями принесли счет на 19 долларов (считая чаевые),")
print("и вы решили скинуться поровну на четверых. То сколько придется с каждого?")
input("Чтобы узнать нажмите Enter.")
print("19 / 4 =", 19 / 4, "$")

print("\nЕсли четверо пиратов найдут ларчик, в котором 107 золотых монет, и")
print("решат разделить добычу поровну. Тосколько целых монет достанется каждому?")
input("Чтобы узнать нажмите Enter.")
print("107 // 4 =", 107 // 4, "монет")

print("\nЕсли те же самые четверо пиратов разделят те же 107 монет из ларчика поровну.")
print(" То сколько останется ничейных монет?")
input("Чтобы узнать нажмите Enter.")
print("107 % 4 =", 107 % 4, "монеты")

input("\n\nнажмите Enter, чтобы выйти.")
```

```
# Манипуляции с цитатой
# Демонстрирует строковые методы

# цитата из Томаса Уотсона, который в 1943 году был директором IBM
quote = "Думаю, на мировом рынке можно будет продать штук пять компьютеров."

print("Исходная цитата:")
print(quote)

print("\nОна же в верхнем регистре:")
print(quote.upper())

print("\nВ нижнем регистре:")
print(quote.lower())

print("\nКак заголовок:")
print(quote.title())

print("\nС ма-а-аленькой заменой:")
print(quote.replace("штук пять", "несколько миллионов"))

print("\nОбращение")
print(quote.swapcase())

print("\ncapitalize():")
print(quote.capitalize())

print("\nstrip():")
print(quote.strip())

print("\nА вот опять исходная цитата:")
print(quote)

input("\n\nнажмите Enter, чтобы выйти.")

# upper()
# lower()
# swapcase() - регистр всех сиволов обращен
# capitalize() - первая буква в строке прописная, остальные строчные
# title()
# strip()
# replace(что заменить, на что заменить, [, сколько раз])
```
